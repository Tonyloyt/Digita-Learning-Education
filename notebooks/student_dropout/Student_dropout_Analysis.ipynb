## Encoding the caterogies with label encoder

```python
#This function will return the integer labels of all categories and will also map the original string label with the integer label.
def label_encode(cat,string1): 
    lab={}
    lb=LabelEncoder()
    data[cat]=lb.fit_transform(data[string1])
    label=data[cat].to_list()
    name=data[string1].to_list()
    #print(type(name))
    for i in range(len(name)):
        lab[name[i]]=label[i]
    return np.unique(lab),lb
```

## Encoding all the categorical variables 

```python
hour_per_week_inassign,lb_hpw=label_encode("Hour_per_week_wriassignment","How many hours per week do you spend on writing assignments?")
time_to_reach_coll,lb_ttrc=label_encode("time_to_reach college","How much time does it take for you to reach college?")
average_attendence,lb_avgatt=label_encode("Average attendence","How much would you consider as your average attendance throughout all the semesters so far?")
Internet_aval_home,lb_inter=label_encode("Internet at home","Internet availability at home")
lecture_time,lb_lecturetime=label_encode("2 hr straight lecture","Can you sit a lecture for 2 hrs straight?")
submission_label,lb_submission=label_encode("submission on time","Do you do your submissions on time?")
Five_hr_lecture,lb_5lect=label_encode("Five lecture straight,woulf you attend all?","If there are 5 hrs of lectures per day, would you attend all?")
Five_hr_practical,lb_5prac=label_encode("Five hr practical staight,do you attend all","If there are 5 hrs of practicals per day, would you attend all?")
teacher_feedback,lb_teacherfeedback=label_encode("Feedback of teacher","Teacher's Feedback")
preffered_transport,lb_preff_transport=label_encode("preffered transportatin to college","What is your preferred mode of transportation to reach college?")
coaching_class,lb_enrolled_coaching=label_encode("Enrolled to coaching class","Have you enrolled for coaching classes?")
```
